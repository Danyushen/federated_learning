```python
%matplotlib inline
import torch
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
plt.style.use('ggplot')
map_location=torch.device('cpu')
root_dir = os.path.dirname(os.path.dirname(os.path.abspath("federated_learning")))
model_dir_i5 = os.path.join(root_dir, "models_i5")
model_dir_gpu = os.path.join(root_dir, "models_gpu")
```

```python
plt.figure("train", (12, 6))
# plt.subplot(1, 2, 1)

fedavg_kd_r20_niid = torch.load(os.path.join(model_dir_gpu,'fedavg_kd_r20_niid.pth'), map_location=map_location)
fedavg_nkd_r20_iid = torch.load(os.path.join(model_dir_gpu,'fedavg_nkd_r20_iid.pth'), map_location=map_location)
fedavg_kd_r20_iid = torch.load(os.path.join(model_dir_gpu,'fedavg_kd_r20_iid.pth'), map_location=map_location)
fedavg_nkd_r20_niid = torch.load(os.path.join(model_dir_gpu,'fedavg_nkd_r20_niid.pth'), map_location=map_location)
fedc_kd_r20_niid = torch.load(os.path.join(model_dir_gpu,'fedc_kd_r20_niid.pth'), map_location=map_location)
fedc_nkd_r20_iid = torch.load(os.path.join(model_dir_gpu,'fedc_nkd_r20_iid.pth'), map_location=map_location)
fedc_kd_r20_iid = torch.load(os.path.join(model_dir_gpu,'fedc_kd_r20_iid.pth'), map_location=map_location)
fedc_nkd_r20_niid = torch.load(os.path.join(model_dir_gpu,'fedc_nkd_r20_niid.pth'), map_location=map_location)

fedavg_nkd_r20_niid_size = fedavg_nkd_r20_niid["model size"]
fedavg_kd_r20_niid_size = fedavg_kd_r20_niid["model size"]
fedavg_nkd_r20_iid_size = fedavg_nkd_r20_iid["model size"]
fedavg_kd_r20_iid_size = fedavg_kd_r20_iid["model size"]
fedc_nkd_r20_niid_size = fedc_nkd_r20_niid["model size"]
fedc_kd_r20_niid_size = fedc_kd_r20_niid["model size"]
fedc_nkd_r20_iid_size = fedc_nkd_r20_iid["model size"]
fedc_kd_r20_iid_size = fedc_kd_r20_iid["model size"]

X = ['Fedavg non-iid','Fedavg iid','Fedcluster non-iid','Fedcluster iid'] 
KD_data_20ep = [fedavg_kd_r20_niid_size,fedavg_kd_r20_iid_size,fedc_kd_r20_niid_size,fedc_kd_r20_iid_size]
NKD_data_20ep = [fedavg_nkd_r20_niid_size,fedavg_nkd_r20_iid_size,fedc_nkd_r20_niid_size,fedc_nkd_r20_iid_size]

X_axis = np.arange(len(X)) 

plt.bar(X_axis - 0.2, KD_data_20ep, 0.4, label = 'KD') 
plt.bar(X_axis + 0.2, NKD_data_20ep, 0.4, label = 'Non-KD') 
plt.xticks(X_axis, X) 
plt.xlabel("Model") 
plt.ylabel("Model Size") 
plt.title("Model Size 20 epochs") 
plt.legend()
plt.show() 
```

```python
plt.figure("train", (12, 6))

fedavg_nkd_r20_niid_size = fedavg_nkd_r20_niid["communication time"] /20
fedavg_kd_r20_niid_size = fedavg_kd_r20_niid["communication time"] /20
fedavg_nkd_r20_iid_size = fedavg_nkd_r20_iid["communication time"] /20
fedavg_kd_r20_iid_size = fedavg_kd_r20_iid["communication time"] /20
fedc_nkd_r20_niid_size = fedc_nkd_r20_niid["communication time"] /20
fedc_kd_r20_niid_size = fedc_kd_r20_niid["communication time"] /20
fedc_nkd_r20_iid_size = fedc_nkd_r20_iid["communication time"] /20
fedc_kd_r20_iid_size = fedc_kd_r20_iid["communication time"] /20

X = ['Fedavg non-iid','Fedavg iid','Fedcluster non-iid','Fedcluster iid'] 
KD_data_20ep = [fedavg_kd_r20_niid_size,fedavg_kd_r20_iid_size,fedc_kd_r20_niid_size,fedc_kd_r20_iid_size]
NKD_data_20ep = [fedavg_nkd_r20_niid_size,fedavg_nkd_r20_iid_size,fedc_nkd_r20_niid_size,fedc_nkd_r20_iid_size]

X_axis = np.arange(len(X)) 

plt.bar(X_axis - 0.2, KD_data_20ep, 0.4, label = 'KD') 
plt.bar(X_axis + 0.2, NKD_data_20ep, 0.4, label = 'Non-KD') 
plt.xticks(X_axis, X) 
plt.xlabel("Model") 
plt.ylabel("Time") 
plt.title("Time 20 epochs") 
plt.legend() 
plt.show() 
```

```python
plt.figure("train", (12, 6))

fedavg_nkd_r20_niid_count = fedavg_nkd_r20_niid["parameter count"]
fedavg_kd_r20_niid_count = fedavg_kd_r20_niid["parameter count"]
fedavg_nkd_r20_iid_count = fedavg_nkd_r20_iid["parameter count"]
fedavg_kd_r20_iid_count = fedavg_kd_r20_iid["parameter count"]
fedc_nkd_r20_niid_count = fedc_nkd_r20_niid["parameter count"]
fedc_kd_r20_niid_count = fedc_kd_r20_niid["parameter count"]
fedc_nkd_r20_iid_count = fedc_nkd_r20_iid["parameter count"]
fedc_kd_r20_iid_count = fedc_kd_r20_iid["parameter count"]

X = ['Fedavg non-iid','Fedavg iid','Fedcluster non-iid','Fedcluster iid'] 
KD_data_20ep = [fedavg_kd_r20_niid_count,fedavg_kd_r20_iid_count,fedc_kd_r20_niid_count,fedc_kd_r20_iid_count]
NKD_data_20ep = [fedavg_nkd_r20_niid_count,fedavg_nkd_r20_iid_count,fedc_nkd_r20_niid_count,fedc_nkd_r20_iid_count]

X_axis = np.arange(len(X)) 

plt.bar(X_axis - 0.2, KD_data_20ep, 0.4, label = 'KD') 
plt.bar(X_axis + 0.2, NKD_data_20ep, 0.4, label = 'Non-KD') 
plt.xticks(X_axis, X) 
plt.xlabel("Model") 
plt.ylabel("Count") 
plt.title("Parameter count 20 epochs") 
plt.legend() 
plt.show() 
```
